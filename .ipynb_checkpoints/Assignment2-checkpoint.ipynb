## Part 1 - Divorce Rate and Margarine Consumption

```r
data(divorce_margarine)
head(divorce_margarine)
```

```r
ggplot(divorce_margarine, aes(x=divorce_rate_maine, y=margarine_consumption_per_capita))+
    geom_smooth(method = lm) +
    geom_point() +
    xlab("Divorce Rate") + 
    ylab("Margarine Consumption Per Capita") +
    theme_minimal()
```

```r
cor.test(divorce_margarine$divorce_rate_maine, divorce_margarine$margarine_consumption_per_capita)
```

We can conclude that the two variables of margarine consumption and divorce rate are significantly correlated as the p-value is less than 0.05 and the correlation coefficient is very close to 1, which indicates a strong positive correlation.

We have to keep in mind; the fact that there is a correlation doesn't mean that there is causation. So the definitive answer would be that margarine consumption wouldn't result in higher divorce rates. 

## Part 2 - English Vocabulary Test

```r
install.packages('car')
library(car)
```

```r
data(GSSvocab)
head(GSSvocab)
```

```r
#Filter for the year 1978 and remove rows with missing value
GSSvocab1978 <- GSSvocab %>%
    filter(year == 1978)

NAfreeGSSvocab1978 <- na.omit(GSSvocab1978)
```

```r
#Is a person’s score on the vocabulary test (‘vocab’) significantly impacted by their level of education (‘educ’)? 
cor.test(NAfreeGSSvocab1978$vocab, NAfreeGSSvocab1978$educ, method = 'pearson')
```

```r
#Visualize the relationship in a plot
ggplot(NAfreeGSSvocab1978, aes(x=vocab, y=educ))+
    geom_jitter(alpha = 0.5)+
    geom_smooth(mehtod = lm) +
    xlab("Vocabulary") +
    theme_minimal()

ggplot(NAfreeGSSvocab1978, aes(vocab, col = educ))+
    geom_density(alpha = 0.5)+
    xlab("Vocabulary") +
    theme_minimal() + 
    facet_wrap(~educ)
```

#### Explanation
There is a moderate positive relationship between the two variables with a rho = 0.54. The p-value < 0.05 indicates a significant relationship between the variables of vacobulary and education. 

```r
#Build a model
model <- lm(vocab ~ educ, data = NAfreeGSSvocab1978)
summary(model)
```

The model also indicates a significant relationship of education on vocabulary as indicated by the ***

```r
#Whether a person is the native of an English-speaking country (‘nativeBorn’) could potentially have an impact on the size of their vocabulary. 
#Visualize the relationship and add the predictor to the model. Briefly explain the results.
ggplot(NAfreeGSSvocab1978, aes(x=vocab, y=educ, color = nativeBorn))+
    geom_jitter(alpha = 0.5) +
    geom_smooth(method=lm) +
    xlab("Vocabulary") +
    ylab("Education") +
    theme_minimal()

#Make point plot with uncertainty line and have two lines one for native yes and one for native no.

model2 <- lm(vocab ~ educ + nativeBorn, data = NAfreeGSSvocab1978)
summary(model2)
```

#### Explanation
The two stars (**) indicate that whether you are born as a native speaker or not has a significant impact on your english vocabulary.

```r
#Try creating the model and briefly explain the results

model3 <- lm(vocab ~ educ*nativeBorn, data = NAfreeGSSvocab1978)
summary(model3)
```

#### Explanation
A one point increase in the vocabulary test is associated with a 0.42 increase in educational level, whcih still makes up a significant influence. 

Participants that are native born have a 0.95 higher score in the vocabulary test, which is not noted as being significnat. 
Niether is the interaction between education and native/non-native. The coefficient of -0.03 suggests the positive effect on vocabulary from education is a bit weaker for participants that are native born than for those that are not. 

```r
#Which model performs best
anova(model, model2, model3)
```

#### Explanation

Adding the interaction term on the other hand does not seem to further improve the model as Pr(>F) > 0.05
