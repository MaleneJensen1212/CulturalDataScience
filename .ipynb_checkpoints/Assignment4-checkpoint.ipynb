```python
df = pd.read_csv("data.csv", sep="#")
df.head()
```

```python
print(df.head())
print(df.dtypes)  # Check data types of all columns
```

```python
# Ensure 'Date' is a datetime object
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract year from the date
df['Year'] = df['Date'].dt.year
df.head()
```

```python
# Split the 'Genre' column into lists of genres
df['Genre'] = df['Genre'].str.split(',')

# Expand the list of genres into separate rows
df = df.explode('Genre')

# Remove extra whitespace from genre names
df['Genre'] = df['Genre'].str.strip()

# Remove brackets from the 'Genre' column
df['Genre'] = df['Genre'].str.replace('[\[\]]', '', regex=True)

df.head()
```

```python
# Group by Year and Genre, summing Streams
genre_year_data = df.groupby(['Year', 'Genre'])['Streams'].sum().reset_index()

# Pivot the table for plotting
pivot_data = genre_year_data.pivot(index='Year', columns='Genre', values='Streams').fillna(0)
```

```python
import matplotlib.pyplot as plt

# Stacked bar plot
pivot_data.plot(kind='bar', stacked=True, figsize=(12, 6))

# Add labels and legend
plt.title('Streams by Genre per Year')
plt.xlabel('Year')
plt.ylabel('Total Streams')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
