# Exercise 1 - introducing correlations 

Let's look at the relationship between FERTILITY and LIFE EXPECTANCY. For this we'll use the gapminder dataset

```r
install.packages('dplyr')
install.packages('ggplot2')
install.packages('dslabs')
```

```r
library(dplyr)
library(ggplot2)
library(dslabs)
```

```r
data(gapminder)
```

Let's restrict our question a bit, and start by looking just at the year 2000

```r
dat2000 <- gapminder %>%
    filter(year==2000)
```

Let's start by just looking at the correlation. The best way to start is to visualise the relationship

```r
ggplot(dat2000,
       aes(x=fertility,
           y=life_expectancy)) +
  geom_point()
```

```r
# guess the correlation. What correlation co-efficient will this have?
```

```r
cor.test(dat2000$life_expectancy, dat2000$fertility)
```

Let's look at another question. What about the relationship between population and life expectancy?

Let's start by just looking at the correlation. The best way to start is to visualise the relationship

```r
ggplot(dat2000,
       aes(x=population,
           y=life_expectancy)) +
  geom_point()
```

This plot is hard to read due to the two very high data points, let's censor them

```r
ggplot(dat2000,
       aes(x=population,
           y=life_expectancy)) +
  geom_point() +
  xlim(0,2e8)
```

```r
# guess the correlation. What correlation co-efficient will this have?
```

```r
cor.test(dat2000$life_expectancy,dat2000$population)
```

Let's apply linear models to both of these relationships

```r
fertility_model <- lm(life_expectancy~fertility, dat2000)
summary(fertility_model)
```

```r
# Questions for reflection:
# How does this relate to the correlation result (cor.test)?
# what is the relationship between R in the correlation and R squared?
# what does R squared mean in this context?
# What does the intercept mean in this model?
# What is the effect of fertility on life expectancy?
```

```r
# do the same yourself for the relationship between life expectancy and population
```

```r
popn_model <- lm(life_expectancy~population,dat2000)
summary(popn_model)
```

# Exercise 2 - interpreting linear models  

Let's go back to last week. We had a question about the influence of number of cylinders on fuel economy. Let's load the data and return to that question

```r
install.packages('fueleconomy')
```

```r
library(fueleconomy)
```

```r
data(vehicles)
```

Let's isolate the relevant data again for Audi and VW midsizes

```r
Audi_VW <- vehicles %>% 
  filter(make=="Audi" & class=="Midsize Cars" | make=="Volkswagen" & class=="Midsize Cars")
```

```r
# remember our question - why is a t.test between makes a bad test?
# remember the plots that were relevant to the question
```

```r
ggplot(data=Audi_VW,
       aes(x=cyl,
           y=cty)) +
  geom_point(position = 'jitter') +
  facet_wrap(~make)
```

Let's use linear models to investigate the effect of number of cylinders on fuel economy in both makes

```r
summary(lm(cty~cyl, 
           data = subset(Audi_VW, make=="Volkswagen")))
```

```r
summary(lm(cty~cyl,
           data = subset(Audi_VW, make=="Audi")))
```

```r
# Questions for reflection:
# are the models any good? How much variance do they explain?
# what are the effects of number of cylinders within each make?
# What are the intercepts for the two models? What do these mean?
# What can we conclude about VW and Audi mid size cars?
```

Let's look at this INTERACTION between variables WITHIN THE MODEL

```r
summary(lm(cty~cyl*make, Audi_VW))
```

```r
# Questions for reflection:
# Is this model better or worse than the other two models? (hint R squared)
# What can we conclude from this model? Look at the parameters. 
# what do the p-values mean in this context?
```

Let's go back to this question about which model is better. We can do this using model comparison. Let's build a model without make, and one with, and see what we learn

```r
m1 <- lm(cty~cyl,Audi_VW)
m2 <- lm(cty~cyl+make,Audi_VW)
m3 <- lm(cty~cyl*make,Audi_VW)
anova(m1,m2,m3)
```

```r
# here we are doing an F test to compare models and find the best one. 
# An F-test is like a t-test. We calculate a test statistic, and use it to find a p-value
# instead of the difference between means divided by standard deviation, we calculate 
# the F statistic as the variance explained by the model, divided by the total variance
# then we find the p value by calculating the area under the curve on an F distribution
# For details, see the Field chapter
```

# Exercise 3 - non-linear predictors

Lets look at the relationship infant mortality and life expectancy in the year 2000 what do you notice?

```r
ggplot(dat2000,
       aes(x=infant_mortality,
           y=life_expectancy)) +
  geom_point()
```

```r
dat2000$infant_mortality2 <- dat2000$infant_mortality^2
```

```r
m1 <- lm(life_expectancy~infant_mortality,dat2000)
m2 <- lm(life_expectancy~infant_mortality+infant_mortality2,dat2000)
```

```r
anova(m1,m2)
```
