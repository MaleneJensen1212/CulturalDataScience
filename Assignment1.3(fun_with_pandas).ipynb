## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
```

```python
#Combine dataframes ranging from the 1950's to the 2020's 
df_50 = rawdata[0]
df_60 = rawdata[1]
df_70 = rawdata[2]
df_80 = rawdata[3]
df_90 = rawdata[4]
df_00 = rawdata[5]
df_10 = rawdata[6]
df_20 = rawdata[7]

merge_df = pd.concat([df_50, df_60, df_70, df_80, df_90, df_00, df_10, df_20])
merge_df
```

```python
merge_df['Launch date'] = pd.to_datetime(merge_df['Launch date']) # Convert the 'Launch date' to datetime format

merge_df['Launch year'] = merge_df['Launch date'].dt.year # Extract the year from the 'Launch date' column

missions_per_year = merge_df.groupby('Launch year').size() # Group by the 'Launch year' and count the number of missions for each year

# Find the year with the most missions
most_missions_year = missions_per_year.idxmax()
most_missions_count = missions_per_year.max()

# Results in text
print(f"The year with the most missions launched is {most_missions_year} with {most_missions_count} missions.")
```

```python
#Most common month for launching - did the exact same but for month
merge_df['Launch month'] = merge_df['Launch date'].dt.month # Extract the month from the 'Launch date' column

missions_per_month = merge_df.groupby('Launch month').size() # Group by the 'Launch month' and count the number of missions for each month

# Find the month with the most missions
most_missions_month = missions_per_month.idxmax()
most_missions_count_month = missions_per_month.max()

# Display the results in text
print(f"The month with the most missions launched is the {most_missions_month}th month with {most_missions_count_month} missions.")
```

```python
ranked_months = missions_per_month.sort_values(ascending=False) #Ranks the months from most launches to fewest launches

# Display the ranked months
ranked_months
#Where the first number is the month and second is number of launches. 
```

## Supervillain challenge

1. Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's
2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade
3. Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table (pandas dataframe)

```python
rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
```

```python
df_30 = rawdata[3]
df_40 = rawdata[4]
df_50 = rawdata[5]
df_60 = rawdata[6]
df_70 = rawdata[7]
df_80 = rawdata[8]
df_90 = rawdata[9]
df_00 = rawdata[10]
df_10 = rawdata[11]

merge_df = pd.concat([df_30, df_40, df_50, df_60, df_70, df_80, df_90, df_00, df_10])
merge_df
```

```python
# Splitting by space and taking only the first part (the year) - as some of the data has a month attached as well
merge_df['Year Debuted'] = merge_df['Year Debuted'].apply(lambda x: x.split(' ')[0])
merge_df
```

```python
merge_df['Year Debuted'] = pd.to_numeric(merge_df['Year Debuted'], errors='coerce') # Convert 'Year Debuted' to numeric, coercing errors to NaN

merge_df = merge_df.dropna(subset=['Year Debuted']) # Drop rows where 'Year Debuted' is NaN (optional depending on your needs)

merge_df.loc[:, 'Year Debuted'] = merge_df['Year Debuted'].astype(int) # Convert the 'Year Debuted' column to integers

merge_df.loc[:, 'Decade'] = (merge_df['Year Debuted'] // 10) * 10 # Create a new column for the decade

decade_counts = merge_df.groupby('Decade').size().reset_index(name='villain_count') # Group by decade and count the number of villains in each decade

decade_counts = decade_counts.sort_values(by='villain_count', ascending=False) # Sort the counts from most to least villains

print(decade_counts)
```

```python
#Write a script that ranks the different comics companies in terms of how many supervillains they have, 
#and display the results in a nice table (pandas dataframe)
#Use merge_df
unique_villain_count = merge_df.groupby('Company')['Character / Team'].nunique().reset_index(name='unique_villain_count') # Group by Company and count the number of unique characters

unique_villain_count_sorted = unique_villain_count.sort_values(by='unique_villain_count', ascending=False) # Sort the df by 'unique_villain_count' in descending order

print(unique_villain_count_sorted)
```

## Stroop challenge

Every year between 2015 and 2021, the students in my Language, Cognition, and the Brain course participated in a version of the Stroop task. Using a stopwatch (ok, using their phones), they recorded how fast they could say a list of things (either reading or naming colors or color words). The column names mean "Reading with No Interference", "Naming with Interference", "Naming with No Interference", and "Reading with Interference". The times are in seconds.

### Stroop challenge 1: 
Transform these data from wide format to long format, so that the result is a dataframe with
- 1 column named "Participant_id" with a unique number for each participant (you can use the row indices)
- 1 column named "Year" with the year data
- 1 column named "Task" that shows which task they were doing
- 1 column named "RT" that shows their response time

```python
# Starter code 1...
df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/Stroop/master/Stroop-raw-over-the-years.csv")
df.head()
```

```python
# Starter code 2...

# Make a new column using the dataframe indices as particpant numbers

df.index.name = 'Participant_id'
df = df.reset_index()

#df.reset_index(inplace = True)
# NOTE: This line does exactly the same thing as the line above:
# it replaces the original df with a new df with the updated index. That's what
# "inplace = True" means. Or, you can just assign the dataframe with the updated index
# to a new dataframe with the same name as the old dataframe, which is what I did above.
# The end result is the same.

df
```

```python
#Making dataframe into long format
df2 = pd.melt(df,
                  id_vars=['Participant_id', 'Year'],  # columns to keep from old df
                  value_vars=['Reading_NoInt', 'Naming_Int', 'Naming_NoInt', 'Reading_Int'],  # melt the task and reaction time information into stacked rows
                  var_name='Task',  #Task are the stored task names
                  value_name='RT') #RT are the corresponsing values

print(df2)
```

## Stroop challenge 2 (Advanced!!!):

Make a new dataframe which shows the mean response time (in seconds) for each task for each year.

```python
new_df = df2.groupby('Task')['RT'].mean().reset_index() #Calculaing the mean for each of the four task groupings. 

# rename column
new_df.rename(columns={'RT': 'Mean_RT'}, inplace=True)
print(new_df)
```
